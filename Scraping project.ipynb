```python
#Define function to call robots.txt
def get_page(url):
    """Scrapes a URL and returns the HTML source.
    
    Args:
        url (string): Fully qualified URL of a page.
    
    Returns:
        soup (string): HTML source of scraped page.
    """

    response = urllib.request.urlopen(urllib.request.Request(url, 
                                                             headers={'User-Agent': 'Mozilla'}))
    soup = BeautifulSoup(response, 
                         'html.parser', 
                         from_encoding=response.info().get_param('charset'))
    
    return soup
```

```python
def get_data(url)->list:
    from random import randint
    diver = []
    diver2 = []
    setoptions = ChromeOptions()
    setoptions.add_argument('--headless')
    driver = webdriver.Chrome()
    driver.get(url)
    numberelem = driver.find_element(By.ID,"load-more-index-tag")
    pagenumbering = int(numberelem.get_attribute("data-page"))
    print(pagenumbering)
    for i in range(2,10000):
        if i == pagenumbering:
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight)")
            elementtoclick = ui.WebDriverWait(driver,randint(5,50)).until(EC.element_to_be_clickable((By.ID,"load-more-index-tag")))
            elementtoclick.click()
            time.sleep(randint(2,5))
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight)")
            numberelem = driver.find_element(By.ID,"load-more-index-tag")
            pagenumbering = int(numberelem.get_attribute("data-page"))
            print(pagenumbering)
    divel = driver.find_elements(By.CLASS_NAME,"artikel")
    for dive in divel:
       diver.append(dive.find_elements(By.CSS_SELECTOR,'a'))
    for dive in diver:
        for div in dive:
            di = div.get_attribute('href')
            diver2.append(di)
    diver2 = [diven for diven in diver2 if len(diven) > 60]
    diver2 = list(set(diver2))
    return diver2
    driver.quit()
```

```python
import nltk
import re

from nltk.tokenize import RegexpTokenizer
from nltk.stem.wordnet import WordNetLemmatizer
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
factory = StemmerFactory()
stemmer = factory.create_stemmer()
my_file = open("C:/Users/zrmuc/Downloads/combined_stop_words.txt", "r")
content = my_file.read()
stopwords = content.split("\n")
def preprocess(sentence):
    sentence = str(sentence)
    sentence = sentence.lower()
    sentence=sentence.replace('{html}',"") 
    cleanr = re.compile('<.*?>')
    cleantext = re.sub(cleanr, '', sentence)
    rem_url=re.sub(r'http\S+', '',cleantext)
    rem_num = re.sub('[0-9]+', '', rem_url)
    tokenizer = RegexpTokenizer(r'\w+')
    tokens = tokenizer.tokenize(rem_num)  
    filtered_words = [w for w in tokens if len(w) > 2 if not w in stopwords]
    stem_words=[stemmer.stem(w) for w in filtered_words]
    return " ".join(stem_words)


#df_merge['cleanText']=df_merge['Text'].map(lambda s:preprocess(s)) 
```
